```python
random.seed(1998)
split_set = set()
result_list = []
for row in tqdm(df.itertuples()):
    split_set.add(row.split)
    if row.split != 'train':
        continue
    data = dict(id=os.path.splitext(row.image_path)[0],
                image=row.image_path,
                domain={"chest_xray": False,"mri": False,"ct_scan": False,"histology": True,"gross": False},
                conversations=[],
               )
    first_flag = True
    if row.not_histology > 0: # 非病理图像
        contents = ['is_histology', 'description']
        random.shuffle(contents)
        for content in contents:
            q = None
            if content == 'is_histology':
                q, a = random.choice(list(zip(is_histology_prompts, assistant_histology_no_prompts)))
            elif content == 'description':
                a = row.caption
                q = random.choice(brief_description_prompts if len(a.split()) < 23 else detailed_description_prompts)
            if q is not None:
                if first_flag:
                    first_flag = False
                    q_list = [q, '<image>']
                    random.shuffle(q_list)
                    q = '\n'.join(q_list)
                data['conversations'].extend([{'from': 'human', 'value': q}, {'from': 'gpt', 'value': a}])
    else:
        contents = ['is_histology', 'description', 'tissue', 'magnification', 'roi']
        random.shuffle(contents)
        for content in contents:
            q = None
            if content == 'is_histology':
                q, a = random.choice(list(zip(is_histology_prompts, assistant_histology_yes_prompts)))
            elif content == 'description':
                a = row.caption
                q = random.choice(brief_description_pathology_prompts if len(a.split()) < 23 else detailed_description_pathology_prompts)
            elif content == 'tissue' and isinstance(row.pathology, str):
                pathology = ast.literal_eval(row.pathology)
                if random.random() < 0.8:
                    q = random.choice(tissue_type_human_prompts)
                    a = random.choice(tissue_type_assistant_responses).format(pathology[0], pathology[1 if len(pathology)>1 else 0], pathology[-1])
                else:
                    q = random.choice(top_three_tissue_types_human_prompts)
                    a = random.choice(top_three_tissue_types_assistant_responses).format(pathology[0], pathology[1 if len(pathology)>1 else 0], pathology[-1])
            elif content == 'magnification' and not math.isnan(row.magnification):
                q, a = random.choice(list(zip(magnification_level_prompts, magnification_level_prompts_assistant_responses)))
                mag_str_list = [['about 0-10x', '0-10x', 'about 1-10x', '1-10x'],
                                ['about 10-20x', '10-20x', '20x'],
                                ['about 20-40x', '20-40x', '40x']]
                a = a.format(random.choice(mag_str_list[int(row.magnification)]))
            elif content == 'roi' and isinstance(row.roi_text, str) and len(ast.literal_eval(row.roi_text))>0:
                q = random.choice(roi_prompts)
                a = row.roi_text
            if q is not None:
                if first_flag:
                    first_flag = False
                    q_list = [q, '<image>']
                    random.shuffle(q_list)
                    q = '\n'.join(q_list)
                data['conversations'].extend([{'from': 'human', 'value': q}, {'from': 'gpt', 'value': a}])
    result_list.append(data)

# with open('/medical-data/zsxm/codes/LLaVA-Med/data/alignment/quilt1m_train.json', 'w') as file:
#     json.dump(result_list, file, indent=2)
print(len(result_list), split_set)
```
